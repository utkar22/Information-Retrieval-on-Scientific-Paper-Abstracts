```python
import os
path='/content/drive/MyDrive/IR_Assignments/A1/Processed_Dataset'
dataset = os.fsencode(path)
```

Creating Unigram Inverted Index

```python
UNIGRAM={}
```

```python
# Creating DocIDs corresponding to Document Names.

UNIGRAM={} #Clearing previous UNIGRAM
idx=0
for doc in os.listdir(dataset):

  doc_name = os.fsdecode(doc) 

  idx=int(doc_name[len('cranfield'):])

  path_to_doc = path + f'/{doc_name}'

  f = open(path_to_doc, 'r')

  words=f.read()[:-1]

  words_list=words.split()

  
  for word in words_list:
    

    if word not in UNIGRAM:
        UNIGRAM[word]=[]   

    if idx in UNIGRAM[word]:
      continue

    UNIGRAM[word].append(idx)
    UNIGRAM[word]=sorted(UNIGRAM[word])
    
    

#Sorting the dictionary according to it's terms

sorted_terms=sorted(UNIGRAM)

UNIGRAM= { i: UNIGRAM[i] for i in sorted_terms}
```

Printing The Unigram Inverted Index

```python
for key in UNIGRAM.keys():
  print(key,'=>',UNIGRAM[key])
```

Saving the Unigram Inverted Index

```python
import pickle
destination_file='/content/drive/MyDrive/IR_Assignments/A1/UNIGRAM.pkl'
```

```python
with open(destination_file, 'wb') as f:
    pickle.dump(UNIGRAM, f)
```
