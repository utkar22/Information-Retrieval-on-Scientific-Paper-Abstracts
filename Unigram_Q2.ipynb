**OPERATON FUNCTIONS**

AND Operation

```python
def AND(list1, list2):
  comp=0

  if list1==[] or list2==[]:
    return [],comp

  sol=[]

  idx1=0
  idx2=0

  while idx1<len(list1)  and idx2< len(list2):
    comp+=1

    if list1[idx1]==list2[idx2]:
      sol.append(list1[idx1])
      idx1+=1
      idx2+=1
    
    elif list1[idx1]<list2[idx2]:
      idx1+=1
    
    else:
      idx2+=1

  return sol,comp
```

OR Operation

```python
def OR(list1, list2):
  comp=0

  sol=[]

  idx1=0
  idx2=0

  while idx1<len(list1)  and idx2< len(list2):
    comp+=1

    if list1[idx1]==list2[idx2]:
      sol.append(list1[idx1])
      idx1+=1
      idx2+=1
    
    elif list1[idx1]<list2[idx2]:
      sol.append(list1[idx1])
      idx1+=1
    
    else:
      sol.append(list2[idx2])
      idx2+=1
  
  if idx1<len(list1):
    sol.extend(list1[idx1:])
  
  if idx2< len(list2):
    sol.extend(list2[idx2:])    

  return sol,comp
```

AND NOT Operation

```python
def AND_NOT(list1, list2):
  comp=0

  if(list1==[]):
    return [],comp
  
  if(list2==[]):
    return list1,comp

  sol=[]

  idx1=0
  idx2=0
  

  while idx1<len(list1)  and idx2< len(list2):
    comp+=1

    if list1[idx1]==list2[idx2]:      
      idx1+=1
      idx2+=1
    
    elif list1[idx1]<list2[idx2]:
      sol.append(list1[idx1])
      idx1+=1
    
    else:
      idx2+=1
    
  if idx1<len(list1):
    sol.extend(list1[idx1:])

  return sol,comp
```

OR NOT Operation

```python
def OR_NOT(list1,lst2):

  All_IDs= doc_mapping.keys()
  list2=[]

  if(lst2==[]):
    list2=All_IDs
  else:    
    
    for id in All_IDs:
      if id not in lst2:
        list2.append(id)  

  return OR(list1, list2)
```

**Computing Query**

```python
#Mapping command to function

def solve(list1, list2, command):
  if command=='AND':
    return AND(list1,list2)

  elif command=='OR':
    return OR(list1, list2)

  elif command=='OR NOT':
    return OR_NOT(list1, list2)
  
  elif command=='AND NOT':
    return AND_NOT(list1, list2)
```

```python
#Solving generalised query

def compute(words, commands):
  sol=[]
  total=0
  query=''

  t1=words[0]
  t2=words[1]

  if(t1 not in UNIGRAM):
    list1=[]
  else:
    list1=UNIGRAM[t1]
  
  if(t2 not in UNIGRAM):
    list2=[]
  else:
    list2=UNIGRAM[t2]
  
  sol,comp=solve(list1,list2,commands[0])
  query+=t1+' '+commands[0]+' '+t2

      
  total+=comp

  for i in range(2,len(words)):

    t=words[i]

    if(t not in UNIGRAM):
      lst=[]
    else:
      lst=UNIGRAM[t]
    
    sol,comp=solve(sol,lst,commands[i-1])
    
    total+=comp
    query+=' '+commands[i-1]+' '+t

  return query, sol, total  
```

```python
#Solving preprocessed query

def simpler_query(input, command_sentence):

  words=[i.strip()  for i in input.split() if(i.strip()!='')]

  commands=[i.strip().upper()  for i in command_sentence.split(',') if(i.strip()!='')]

  return compute(words,commands)
```

```python
#Function to process query

def query(no, input, command_sentence):

  # process this input from user

  query, ans, comparisions= simpler_query(input,command_sentence)
  docNames=''
  for id in ans:
    docNames+=doc_mapping[id]+' '

  print('Query '+str(no)+' : '+ query)
  print('Number of documents retrieved for query ' +str(no)+' : '+ str(len(ans)))
  print('Names of the documents retrieved for query '+ str(no)+ ' : '+ docNames)
  print('Number of comparisons required for query '+ str(no)+' : '+ str(comparisions))
```

**Running The Program**

```python
def main():
  test_cases=int(input())

  for i in range(1, test_cases+1):
    sentence=input()
    commands=input()
    query(i,sentence,commands)
```
