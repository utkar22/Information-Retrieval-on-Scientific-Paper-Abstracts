```python
import os
path='/content/drive/MyDrive/IR_Assignments/A1/Processed_Dataset'
dataset = os.fsencode(path)
```

Creating Bigram Inverted Index

```python
BIGRAM={}
```

```python
# Creating DocIDs corresponding to Document Names.

BIGRAM={} #Clear the previous BIGRAM
idx=0
for doc in os.listdir(dataset):
  
  doc_name = os.fsdecode(doc) 

  idx=int(doc_name[len('cranfield'):])

  path_to_doc = path + f'/{doc_name}'

  f = open(path_to_doc, 'r')

  words=f.read()[:-1]
  

  words_list=words.split()

  
  

  for i in range(len(words_list)-1):

    word=words_list[i]+' '+words_list[i+1]

    if word not in BIGRAM:
        BIGRAM[word]=[]   

    if idx in BIGRAM[word]:
      continue

    BIGRAM[word].append(idx)
    BIGRAM[word]=sorted(BIGRAM[word])

    
    

#Sorting the dictionary according to it's terms

sorted_terms=sorted(BIGRAM)

BIGRAM= { i: BIGRAM[i] for i in sorted_terms}
```

Printing The Bigram Inverted Index

```python
for key in BIGRAM.keys():
  print(key,'=>',BIGRAM[key])
```
